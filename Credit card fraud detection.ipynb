```python
# distribution of legit transaction and fraudulent transaction
data['Class'].value_counts()
```

```python
# this dataset is highly unbalenced
# 0 - Normal transaction
# 1 - Fraudulent transaction
```

```python
# separate the data for analysis
legit = data[data.Class == 0 ]
fraud = data[data.Class == 1 ]
```

```python
legit.shape
```

```python
fraud.shape
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compaire the values of both of the classes
data.groupby('Class').mean()
```

```python
# under sampling
legit_sample = legit.sample(492)
```

```python
new_data=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_data.head()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
x = new_data.drop(columns = 'Class',axis=1)
y = new_data['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
# split the data into train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#model training 
# Logistic Regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
# training the Logistic Regression model
model.fit(x_train,y_train)
```

```python
# Evaluation
#accuracy on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print('accuracy on training data:',training_data_accuracy)
```

```python
# accuracy on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy score on test data:',test_data_accuracy)
```
